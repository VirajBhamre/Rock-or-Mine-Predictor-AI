```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df=pd.read_csv("sonar data.csv",header=None)
```

```python
df
```

```python
df[60]=df[60].replace({"R":0,"M":1})
```

```python
x=df.drop(60,axis=1)
y=df[60]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y.sample(5)
```

```python
model=keras.Sequential([
    layers.Dense(64,input_shape=(60,),activation='relu'),
    layers.Dense(128,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    loss="binary_crossentropy",
    metrics=["accuracy"],
    optimizer='adam'
)
```

```python
model.fit(x_train,y_train,epochs=50)
```

```python
model.evaluate(x_test,y_test)
```

```python
test_data=pd.read_csv("test_data.csv",header=None)
```

```python
test_data[60]=test_data[60].replace({"R":0,"M":1})
```

```python
x_testing=test_data.drop(60,axis=1)
y_testing=test_data[60]
```

```python
model.evaluate(x_testing,y_testing)
```

```python
model.save("Rock_Or_Mine_Predictor.keras")
```
